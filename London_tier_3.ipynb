**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great!

```python
print(properties_by_year['Hackney'][1998])
print(properties_by_year['Hackney'][2018])
print(properties_by_year['Hackney'][2018] / properties_by_year['Hackney'][1998])
```

```python
def create_price_ratio(borough):
    if borough.title() in properties_by_year:
        price_ratio = properties_by_year[borough.title()][2018] / properties_by_year[borough.title()][1998]
        print(price_ratio)
    else:
        print("This is not a London borough")
```

```python
create_price_ratio('Camden')
```

```python
create_price_ratio('camden')
```

```python
properties_by_year.diff()
```

```python
price_percentage_by_year = ((properties_by_year / (properties_by_year - properties_by_year.diff())) -1) * 100
```

```python
print(price_percentage_by_year)
```

```python
print(price_percentage_by_year[['Hackney', 'Waltham Forest']])
```

```python
from collections import defaultdict
```

```python
twenty_year_ratio = defaultdict(list)

for Borough in properties_by_year.columns:
    ratio = properties_by_year[Borough][2018] / properties_by_year[Borough][1998]
    twenty_year_ratio[Borough].append(ratio)
```

```python
print(twenty_year_ratio)
```

```python
twenty_year_ratio = pd.DataFrame.from_dict(twenty_year_ratio).transpose()
```

```python
sorted_ratio = twenty_year_ratio.sort_values(by=0, ascending= False)
```

```python
print(sorted_ratio)
```
